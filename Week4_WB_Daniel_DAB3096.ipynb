```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

**UPLOADING THE DATASETS**

```python
 # import for interactive visualisation of graphs
import plotly.express as px
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# create the scatter chart using Plotly Express
fig_fertility_expectation_region = px.scatter(data_frame=fd,
                 x='fertility_rate',               # x-axis consists of fertility rate
                 y='Life_Expectancy',              # y-axis consists of life expectancy
                 size='Population',                # bubble size based on population
                 size_max=100,                      # set the maximum size for bubbles
                 hover_name='Country Code',        # will view country names on hovering
                 color='Region',                   # will color accordingly to region
                 animation_frame='Year',           # will animate in time frame
                 animation_group='Country Code',   # animation format
                 template='plotly_dark',           # black as template
                 labels={'Region': 'Region'},      # standardize the label
                 range_x=[0, 10],                  # range of x axis
                 range_y=[10, 90])                 # range of y axis

fig_fertility_expectation_region.update_layout(title='fertility_rate vs Life_Expectancy',        # title of graph
                  xaxis_title='fertility_rate',  # x axis label
                  yaxis_title='Life_Expectancy_Rate')           # y axis label

fig_fertility_expectation_region.show()
```

**Years vs Population**

```python
# plotting Line Chart
years = fd['Year']
population = fd['Population']

# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust size of the figure as needed
plt.plot(years, population, linestyle='-', color='b', label='Population Trends')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Trends Over Years')
plt.grid(True)

# Add legend
plt.legend()

# Show plot
plt.tight_layout()  # Ensures all elements fit nicely in the plot area
plt.show()
```

**Fertility rate distribution**

```python
sb.histplot(fd['fertility_rate'])
```

**Life expectancy variation**


```python
plt.figure(figsize=(12, 8))
sb.lineplot(x='Year', y='Life_Expectancy', data=fd, ci=None)
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy Variation by Country')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # Adjust legend position
plt.tight_layout()
plt.show()
```

**Correlation Analysis**

```python
corr_matrix = fd[['Year', 'Population','fertility_rate', 'Life_Expectancy']].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sb.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

**Regional Analysis**

```python
# Region Vs Life Expectancy Rate
plt.figure(figsize=(10, 6))
sb.barplot(x='Region', y='Life_Expectancy', data=fd, color='g')
plt.xlabel('Region')
plt.ylabel('Average Life Expectancy')
plt.title('Average Life Expectancy by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Region Vs fertility rate
plt.figure(figsize=(10, 6))
sb.barplot(x='Region', y='fertility_rate', data=fd, color='pink')
plt.xlabel('Region')
plt.ylabel('fertility rate')
plt.title('Fertility Rate Vs Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Region Vs Population
plt.figure(figsize=(10, 6))
sb.barplot(x='Region', y='Population', data=fd, color='r')
plt.xlabel('Region')
plt.ylabel('Population')
plt.title('Population Vs Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
